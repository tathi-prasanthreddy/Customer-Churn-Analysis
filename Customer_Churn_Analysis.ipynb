# **Missing Data - Initial Intuition**


Here, we don't have any missing data.

General Thumb Rules:

*   For features with less missing values- can use regression to predict the missing values or fill with the mean of the values present, depending on the feature.

*   For features with very high number of missing values- it is better to drop those columns as they give very less insight on analysis.

*   As there's no thumb rule on what criteria do we delete the columns with high number of missing values, but generally you can delete the columns, if you have more than 30-40% of missing values.


# **Data Cleaning**
1. Create a copy of base data for manupulation & processing

```python
customer_data = churn_data.copy()
```

```python
customer_data.head()
```

2. Total Charges should be numeric amount. Let's convert it to numerical data type

```python
customer_data.TotalCharges = pd.to_numeric(customer_data.TotalCharges,errors = 'coerce')
customer_data.isnull().sum()
```

3. As we can see there are 11 missing values in TotalCharges column.

```python
customer_data.loc[customer_data ['TotalCharges'].isnull() == True]
```

4. Missing Value Treatement

Since the % of these records compared to total dataset is very low ie 0.15%, it is safe to ignore them from further processing.

```python
customer_data.dropna(how = 'any', inplace = True)
```

5. Divide customers into bins based on tenure e.g. for tenure < 12 months: assign a tenure group if 1-12, for tenure between 1 to 2 Yrs, tenure group of 13-24; so on..

```python
labels = ["{0} - {1}".format(i, i + 11) for i in range(1, 72, 12)]

customer_data['tenure_group'] = pd.cut(customer_data.tenure, range(1, 80, 12), right=False, labels=labels)
```

```python
customer_data['tenure_group'].value_counts().reset_index()
```

6. Remove columns not required for processing

```python
customer_data.drop(columns=['customerID','tenure'],axis=1,inplace = True)
```

```python
customer_data.head()
```

Data Exploration

1. Plot distibution of individual predictors by churn

**Univariate Analysis**

```python
for i,predictor in enumerate(customer_data.drop(columns=['Churn', 'TotalCharges', 'MonthlyCharges'])):
    plt.figure(i)
    sns.countplot(data = customer_data,x = predictor,hue = 'Churn')
```

2. Convert the target variable 'Churn' in a binary numeric variable i.e. Yes=1 ; No = 0

```python
customer_data['Churn'] = np.where(customer_data['Churn'] == 'Yes',1,0)
customer_data.head()
```

3. Convert all the categorical variables into dummy variables

```python
customer_data_dummies = pd.get_dummies(customer_data)
customer_data_dummies.head()
```

4. Relationship between Monthly Charges and Total Charges

```python
sns.lmplot(data=customer_data_dummies, x='MonthlyCharges', y='TotalCharges', fit_reg=False)
```

```python
customer_data_dummies['MonthlyCharges'] .corr(customer_data_dummies['TotalCharges'])
```

Total Charges increase as Monthly Charges increase - as expected.

5. Churn by Monthly Charges and Total Charges

```python
Mth = sns.kdeplot(customer_data_dummies.MonthlyCharges[(customer_data_dummies["Churn"] == 0) ],
                color="Red", fill = True)
Mth = sns.kdeplot(customer_data_dummies.MonthlyCharges[(customer_data_dummies["Churn"] == 1) ],
                ax = Mth ,color="Blue", fill = True)
Mth.legend(["No Churn","Churn"],loc='upper right')
Mth.set_ylabel('Density')
Mth.set_xlabel('Monthly Charges')
Mth.set_title('Monthly charges by churn')
```

```python
Tot = sns.kdeplot(customer_data_dummies.TotalCharges[(customer_data_dummies["Churn"] == 0) ],
                color="Red", fill = True)
Tot = sns.kdeplot(customer_data_dummies.TotalCharges[(customer_data_dummies["Churn"] == 1) ],
                ax =Tot, color="Blue", fill= True)
Tot.legend(["No Churn","Churn"],loc='upper right')
Tot.set_ylabel('Density')
Tot.set_xlabel('Total Charges')
Tot.set_title('Total charges by churn')
```

**Surprising insight ** as higher Churn at lower Total Charges

6. Build a corelation of all predictors with 'Churn' **

```python
plt.figure(figsize=(18,6))
customer_data_dummies.corr()['Churn'].sort_values(ascending = False).plot(kind='bar')
```

**Derived Insight: **

HIGH Churn seen in case of Month to month contracts, No online security, No Tech support, First year of subscription and Fibre Optics Internet

LOW Churn is seens in case of Long term contracts, Subscriptions without internet service and The customers engaged for 5+ years

Factors like Gender, Availability of PhoneService and # of multiple lines have alomost NO impact on Churn

This is also evident from the Heatmap below

```python
plt.figure(figsize=(12,12))
sns.heatmap(customer_data_dummies.corr(), cmap="Paired")
```
